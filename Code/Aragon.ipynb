# Text Processing

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import re

# Download stopwords and stemmer
nltk.download('stopwords')
nltk.download('punkt')

# Define the preprocessing function
stop_words = set(stopwords.words('english'))
def preprocess_text(text):
    # Check if text is NaN
    if pd.isna(text):
        return ""

    # Convert to lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'http\S+', '', text)

    # Remove non-alphabetic characters
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Tokenize the text
    words = word_tokenize(text)

    # Remove stop words and punctuation
    words = [word for word in words if word not in stop_words and word.isalpha()]

    # Join the remaining words into a string
    text = ' '.join(words)

    return text

# Apply the preprocessing function to the 'Content' column
ant['Preprocessed'] = ant['Content'].apply(preprocess_text)
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Initialize the sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Convert the 'Date' column to a datetime object and set it as the index

ANT['Date'] = pd.to_datetime(ANT['Date'], errors='coerce')
ANT.set_index('Date', inplace=True)

# Define a function to compute the sentiment score for each day
def daily_sentiment_score(text):
    sentiment = analyzer.polarity_scores(text)
    return sentiment['compound']

# Compute the daily sentiment score using resample and apply
ant_daily = ANT['Preprocessed'].resample('D').apply(lambda x: x.apply(daily_sentiment_score).mean())
```

```python
ant_daily
```

```python
ant_daily_sentiment = ant_daily.to_frame(name='Sentiment Score').reset_index()
```

```python
import plotly.graph_objs as go

# Create a line graph using Plotly
fig = go.Figure()

# Add a line trace to the figure
fig.add_trace(go.Scatter(x=ant_daily_sentiment['Date'], y=ant_daily_sentiment['Sentiment Score'], mode='lines'))

# Customize the layout of the figure
fig.update_layout(title='Aragon Discord Daily Sentiment Score', xaxis_title='Date', yaxis_title='Sentiment Score')

# Display the figure
fig.show()
```

```python
positive_count = len(ant_daily_sentiment[ant_daily_sentiment['Sentiment Score'] > 0])
negative_count = len(ant_daily_sentiment[ant_daily_sentiment['Sentiment Score'] < 0])
```

```python
import plotly.express as px

# Calculate positive and negative sentiment counts
positive_count = (ant_daily_sentiment['Sentiment Score'] > 0).sum()
negative_count = (ant_daily_sentiment['Sentiment Score'] < 0).sum()

# Create a dataframe for histogram
hist_ant = pd.DataFrame({
    'Sentiment': ['Positive', 'Negative'],
    'Count': [positive_count, negative_count]
})

# Create a plotly histogram
fig = px.histogram(hist_ant, x='Sentiment', y='Count', color='Sentiment',
                   title='Aragon Discord Sentiment Distribution')

# Set y-axis label
fig.update_layout(yaxis_title='Count (days)')

# Add count number on each bar
fig.update_traces(texttemplate='%{y}', textposition='outside')

fig.show()
```

```python
# Calculate overall statistics on the sentiment scores
overall_sentiment_stats = ant_daily_sentiment['Sentiment Score'].describe()

# Output overall sentiment statistics
print("Overall Sentiment Statistics:")
print(overall_sentiment_stats)
```
